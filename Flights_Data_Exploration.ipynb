# Flights Data Exploration Challenge¶

Importing the data and loading of data

```python
import pandas as pd
#import numpy as np
#import math as mt
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df_flights=pd.read_csv("flights.csv")
df_flights.head()
```

Cleaning of the data

```python
df_flights.isnull()
```

```python
#Check if there are any null values 
df_flights.isnull().sum()
```

```python
df_flights.DepDel15=df_flights.DepDel15.fillna(0)
df_flights.isnull().sum()
```

```python
df_flights.describe().isnull().sum()
```

Identifying and removing of outliers

```python
def show_stats(var_data):
    min_val=var_data.min()
    max_val=var_data.max()
    mean_val=var_data.mean()
    med_val=var_data.median()
    mod_val=var_data.mode()[0]
    
    print(var_data.name,"\n minimum value:{:} \n maximum value:{:} \n mean value:{:} \n median value:{:} \n mode value:{:}".format(min_val,max_val,mean_val,med_val,mod_val))
    
    
    #Create a figure for 2 subplots
    
    fig,ax=plt.subplots(2,1,figsize=(10,6))
    
    #Plot Histogram
    
    ax[0].hist(var_data)
    ax[0].set_ylabel("Frequency")
    
    ax[0].axvline(x=min_val,color='grey',linestyle='--',linewidth=2)
    ax[0].axvline(x=max_val,color='cyan',linestyle='--',linewidth='2')
    ax[0].axvline(x=mean_val,color='red',linestyle='--',linewidth='2')
    ax[0].axvline(x=med_val,color='yellow',linestyle='--',linewidth='2')
    ax[0].axvline(x=mod_val,color='blue',linestyle='--',linewidth='2')
    
    
    #PLot boxlplot
    
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel("Value")
    
    #Title of the Figure
    
    fig.suptitle(var_data.name)
    
    fig.show()
    
#Calling the Function

delay_fields=['DepDelay','ArrDelay']
for col in delay_fields:
    show_stats(df_flights[col])
```

```python
#Trim outliers from ArrDelay based on 1% and 90%

ArrDelay_01ptile=df_flights.ArrDelay.quantile(0.01)
ArrDelay_90ptile=df_flights.ArrDelay.quantile(0.90)

df_flights=df_flights[df_flights.ArrDelay < ArrDelay_90ptile]
df_flights=df_flights[df_flights.ArrDelay > ArrDelay_01ptile]

for col in delay_fields:
    show_stats(df_flights[col])
    
    
```

```python
#Trim outloiers for DepDelay

DepDelay_01ptile=df_flights.DepDelay.quantile(0.01)
DepDelay_90ptile=df_flights.DepDelay.quantile(0.90)

df_flights=df_flights[df_flights.DepDelay < DepDelay_90ptile]
df_flights=df_flights[df_flights.DepDelay >DepDelay_01ptile]

for col in delay_fields:
    show_stats(df_flights[col])
```

Which departure airport has the highest average departure delay?

```python
Departure_Airport_group=df_flights.groupby(df_flights.OriginAirportName)

mean_departure_delay=pd.DataFrame(Departure_Airport_group["DepDelay"].mean()).sort_values('DepDelay',ascending=False)
print(mean_departure_delay)
mean_departure_delay.plot(kind='bar',figsize=(12,6))
```

so, we can say that Long Beach Airport has the highest average departure delays with a value of -3.447844.

Which route (from origin airport to destination airport) has the most late arrivals?

```python
df_flights.head()
```

```python
routes=pd.Series(df_flights["OriginAirportName"] +">"+df_flights["DestAirportName"])
df_flights=pd.concat([df_flights,routes.rename("Route")],axis=1)
df_flights.head()
```

```python
route_group=df_flights.groupby(df_flights.Route)

sum_arr_delay=pd.DataFrame(route_group['ArrDel15'].sum()).sort_values('ArrDel15',ascending=False)
sum_arr_delay
```

so we can say that Route from origin Airport San Francisco International to Destination Airport Los Angeles International has most numbe rof late arrivals and that is 90.

Which route has the highest average arrival delay?

```python
avg_arr_delay=pd.DataFrame(route_group["ArrDelay"].mean()).sort_values('ArrDelay',ascending=False)
avg_arr_delay.head()
```
